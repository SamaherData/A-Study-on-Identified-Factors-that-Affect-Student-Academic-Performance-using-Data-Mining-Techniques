```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
por = pd.read_csv("C:/Users/DCU/Desktop/Data/student-por.csv",delimiter=';')
mat = pd.read_csv("C:/Users/DCU/Desktop/Data/student-mat.csv", delimiter=';')


por['course'] = 'por'
mat['course'] = 'mat'

combined_data = pd.concat([por, mat], ignore_index=True)
cols = combined_data.columns.tolist()
cols = cols[-1:] + cols[:-1]
combined_data = combined_data[cols]
    
def map_G3_01(x):
    if x<10:
        return 0
    else: return 1

combined_data["G3_01"] = combined_data.G3.apply(map_G3_01)
```

```python
cat_vars= combined_data.select_dtypes(include=['object'])
dummies_data = pd.DataFrame(combined_data,copy=True)
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(dummies_data[var], prefix=var,drop_first=True)
    data1=dummies_data.join(cat_list)
    dummies_data=data1
    
data_vars=dummies_data.columns.values.tolist()
to_keep=[i for i in data_vars if i not in cat_vars]

dummies_data_final=dummies_data[to_keep]
#(Li, 2017)
```
