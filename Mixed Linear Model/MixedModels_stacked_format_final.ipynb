```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
por = pd.read_csv("C:/Users/test/Desktop/data/student-por.csv",delimiter=';')
mat = pd.read_csv("C:/Users/test/Desktop/data/student-mat.csv",delimiter=';')
por['course'] = 'por'
mat['course'] = 'mat'

combined_data = pd.concat([por, mat], ignore_index=True)
cols = combined_data.columns.tolist()
cols = cols[-1:] + cols[:-1]
combined_data = combined_data[cols]
n_combined_data = pd.DataFrame(combined_data,copy=True)
n_combined_data["id"] = n_combined_data.index
cols = n_combined_data.columns.tolist()
cols = cols[-1:] + cols[:-1]
n_combined_data = n_combined_data[cols]

```

```python
cat_vars= n_combined_data.select_dtypes(include=['object']).columns.to_list()
dummies_data = pd.DataFrame(n_combined_data,copy=True)
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(dummies_data[var], prefix=var, drop_first=True)
    data1=dummies_data.join(cat_list)
    dummies_data=data1
    
data_vars=dummies_data.columns.values.tolist()
to_keep=[i for i in data_vars if i not in cat_vars]

dummies_data_final=dummies_data[to_keep]
```

```python
#(Ponomaroff, 2018)
long_data = pd.wide_to_long(dummies_data_final,stubnames='G',i=['id'],j='Grade')
long_data.sort_index(inplace=True)
```
